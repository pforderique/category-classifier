# Category Classifier Walkthrough

This notebook mirrors the CLI flow: load data, inspect labels, train, and benchmark.

````python
from pathlib import Path

from category_classifier.dataset import load_transactions
from category_classifier.encoder import SentenceTransformerEncoder
from category_classifier.training import TrainConfig, train_model
from category_classifier.artifacts import save_model_pack
````

````python
data_path = Path("../data/transactions.tsv")
df = load_transactions(data_path)
df.head()
````

````python
encoder = SentenceTransformerEncoder("sentence-transformers/paraphrase-MiniLM-L3-v2", device="cpu")
config = TrainConfig(epochs=40, batch_size=64, seed=42)
result = train_model(df=df, encoder=encoder, model_name="personal-v1", device="cpu", config=config)
result.metrics
````

````python
pack_path = save_model_pack(Path("../artifacts/personal-v1"), result)
pack_path
````

You can benchmark from terminal:

```bash
category-classifier benchmark --model-pack personal-v1 --artifacts-dir ../artifacts --devices cpu,mps
```
